```python
boleta=input("Ingrese tipo de boleta GENERAL/VIP: ").upper() #.upper se utiliza para pasar las variables de caracter a todo mayuscula y asi hacer mas facil el condicional
edades=int(input("Ingrese la edad: ")) #Ejercicio de precio de boletas para entrar a un lugar
if boleta=="GENERAL":
  if edades<12:
     print("El precio es de $8000")
  elif 12<= edades <=17:
     print("El precio es de $10000")
  elif 18<= edades<=59:
     print ("El precio es de $15000")
  elif edades>=60:
     print ("El precio es de $7000")
  else:
     print("Edad no válida")
elif boleta=="VIP":
  if 0<edades<12:
     print("El precio es de $12000")
  elif 12<=edades<=17:
    print("El precio es de $15000")
  elif 18<=edades<60:
    print ("El precio es de $20000")
  elif 60<=edades:
    print ("El precio es de $12000")
  else:
    print("Edad no válida")
else:
  print("Tipo de boleta no valido")
```

```python
numero=int(input("Ingrese un numero: "))# Ejercicio de numero par y divisible por 10
if numero==0:
  print("Numero especial")
elif numero % 2 == 0 and numero % 10 == 0:
  print("El número es par y divisible por 10")
else:
  print("No cumple la condición ")
```
