```python
N=int(input("¿Que cantidad de números va a ingresar?: "))
N_pares=0
N_impares=0
T_numeros=0
for i in range (N):
  numero=int(input("Ingrese un número: "))
  T_numeros+=numero
  if numero%2==0:
    N_pares+=numero
  else:
    N_impares+=numero
print("La suma de los números pares es: ",N_pares)
print("La suma de los números impares es: ",N_impares)
print("El total de la suma de los números ingresados es: ",T_numeros)
```

```python
Saldo=1000
while True:
 print("A. Consultar saldo")
 print("B. Retirar dinero")
 print("C. Depositar dinero")
 print("D. Salir")

 Opcion=input("Eliga una opción: ").upper()
 if Opcion=="A":
  print(f"El saldo es {Saldo}")
 elif Opcion=="B":
  Retiro=int(input("Cuanto va a retirar: "))
  if Retiro>Saldo:
   print("No hay ese dinero en este cajero")
  elif Retiro<=Saldo:
    print("Retiro exitoso")
    Saldo-=Retiro
 elif Opcion=="C":
  Deposito=int(input("Cuanto dinero va a depositar: "))
  print("Deposito exitoso")
  Saldo+=Deposito
 elif Opcion=="D":
  break
```

```python
N=int(input("Cuantas medidas tomó: "))
lista=[]
Suma=0
for i in range(1,N+1):
  Numero=float(input(f"Ingrese la medida {i}: "))
  lista.append(Numero)
promedio=sum(lista)/len(lista)
suma=0
for elemento in lista:
  suma+=desvia(elemento,promedio,N)
desviacion=(suma)**0.5
print(f"El promedio de los numeros ingresados es: {promedio}")
print(f"La desviación estandar de los numero es: {desviacion}")
```
