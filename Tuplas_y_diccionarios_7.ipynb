```python
diccionario_personas={"Nombre":["Juan","Camilo","Lessa"],"Edad":[22,18,20],"Ciudad":["Cali","Barranquilla"]}
```

```python
diccionario_personas["Ciudad"]
```

```python
diccionario_personas["Ciudad"].append("medellin")#Forma de añadir otros valores a un diccionario
```

```python
diccionario_personas["profesion"]=["Chef","Arquitecto","Ingenieria"]#Forma e añadir una una clave con su valor
print(diccionario_personas)
```

```python
diccionario_personas["Nombre"][2]="Laura" #Forma de cambiar un elemento de los valores
print(diccionario_personas)
```

```python
print(diccionario_personas.keys())
print(diccionario_personas.values())
```

```python
for key, value in diccionario_personas.items():#Forma de ver las claves y su valor
  print(f"La llave es  {key} y su valor es {value}")
```
