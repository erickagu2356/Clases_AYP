```python
def resta(a,b):
  c=a-b
  return c #Para devolver la variable que necesite
```

```python
resultado= resta(8,3)#Hay que tener cuidado con el orden porque si no se insertan bien los elemnetos puede arrojar un resultado diferente
print(resultado)#Si yo inserto 1 o mas de 2 vaores en la resta da error
```

```python
def resta1(a,b,c=0):#Si yo le asigno un numero a una variable ya no dara error al no escriirla cuando la pidan
  return a-b+c
```

```python
resultado1=resta1(a=1,b=5)
print(resultado1)
```

```python
def resta2(a=0,b=0,c=0):
  return a-b+c,c-a-b#Si hago dos procedimientos y los separo con una coma el me dara una tupla con 2 resultados
```

```python
resta2(1,2, 3)#Ejemplo
```

```python
resulta1,resulta2= resta2(2,3,6)#Forma de aprovechar dos formas sin creara otra funcion para otra operacion
print("El resultado1 es: ",resulta1)
print("El resultado2 es: ",resulta2)
```

```python
def num_mayor(a,b,c):
  d=[a,b,c]
  return max(d)
```

```python
numero_mayor= num_mayor(55,-8,64.3)
print(numero_mayor)
```

```python
def n_pares(lista):
  n_lista=[]
  for i in lista:#Forma de escanear los elementos de la lista
   if i % 2==0:
    n_lista.append(i)
  return n_lista
```

```python
lista=[2,7,9,87,12,45,44,46,48,24,56]
lista_pares= n_pares(lista)
print(lista_pares)
```

```python
def factorial(N):
 num_fac=1
 for i in  range(1,N+1):
  num_fac == num_fac*i
  print(num_fac)
 return num_fac
```
