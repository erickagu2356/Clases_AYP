```python
n=int(input("Ingrese la cantidad de numeros que vaya a ingresar "))#
lista=[]

for i in range(n):
 numero=float(input("Ingrese un número "))
 if numero>5:
  lista.append(numero)
print("El numero maximo en la lista es ", max(lista))
print("El numero minimo en la lista es ", min(lista))
print("El promedio de los numeros de la lista es ",sum(lista)/len(lista))
```
